```python
#instance method
# method vs function

class Vehicle:
  def __init__(self, color):
    self.color=color
    self.vehicle_category='Heavy'

  def category(self, cate_gory='None'):
    if cate_gory !='Null' :
      self.vehicle_category=cate_gory

  def printCategory(self):
    print(self.vehicle_category)

bus=Vehicle('blue')
bus.printCategory()
bus.category('Light')
bus.printCategory()
```
