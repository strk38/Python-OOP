```python
class Calculator:
  def __init__(self):
    self.a=0
    self.b=0
    self.output=0
    print('Lets Calculate')
  


  def add(self,a,b):
    print(a+b)

  def sub(self,a,b):
    print(a-b)
  
  def mul(self,a,b):
    print(a*b)
  
  def div(self,a,b):
    print(a/b)

calc1=Calculator()
a=4
b=2
op='+'
print('Result: ',end='')
if op=='+':
  calc1.add(a,b)

elif op=='-':
  calc1.sub(a,b)
elif op=='*':
  calc1.mul(a,b)
else:
  calc1.div(a,b)

```

### Method overloading (ln 18, 22)

```python
class Customer:
  def __init__(self,name):
    self.name=name

  def greet(self, name=None):
    if name==None:
      print('Hello!')
    else:
      print(f'Hello {name} !')

  def purchase(self, *args):

    print(f'{self.name}, you purchased {len(args)} item(s):')
    for i in args:
      print(i)

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice") 
```

```python
class Cat:
  def __init__(self,color=None, act=None):
    if color==None:
      self.color='White'
      self.act='sitting'
    elif color!=None and act==None:
      self.color=color
      self.act='sitting'
    else:
      self.color=color
      self.act=act

  def printCat(self):
      print(f'{self.color} cat is {self.act}')


  def changeColor(self, color):
    self.color=color

c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
# #   Task 22 ::: VvvvIiii      # #
class Vaccine:
  def __init__(self, vacName, origin, duration):
    self.vacName=vacName
    self.origin=origin
    self.duration=duration

class Person:
  def __init__(self, P_Name, age, ocupation=None):
    self.P_Name=P_Name
    self.age=age

    if ocupation==None:
      self.occupation='General Citizen'
    else:
      self.occupation=ocupation

    self.dose=[]
    self.i=0
    self.vaccine=''
    self.v_duration=0

  def pushVaccine(self, V_name, dose=None):
    if V_name.vacName=='Sinopharm' and self.age<25 and dose==None:
      print(f'Sorry {self.P_Name}, minimum age for this vaccine is 25years for now')
    else:
      self.v_duration=V_name.duration         # #  Accessing variables from another class # #
      if dose==None:
        self.dose.append('1st Dose')
        self.vaccine=V_name.vacName
        self.i+=1
        print(f'{self.dose[0]} done for {self.P_Name}')
      else:
        if V_name.vacName!=self.vaccine :
          print(f"Sorry {self.P_Name}, u cant take different vaccines")
        else:
          self.dose.append(dose)
          self.i+=1
          print(f'{dose} done for {self.P_Name}')
        
  def showDetail(self):
    print('Name ',self.P_Name, " Age: ", self.age," Type: ",self.occupation)
    print("Vaccine name: ", self.vaccine)
    for i in range(len(self.dose)):
      print(self.dose[i]," : Given")
    else:
      if len(self.dose)<2:
        print(f'{len(self.dose)+1}nd dose: Please come after {self.v_duration} days')

astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)                       # # Passing location/Reference of object to Method of another class # #
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
