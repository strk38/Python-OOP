```python
s='OhMyBR@CU20' #'ohmybracu' #'ohMyBR@CU'
lower=False
for i in s:
  if i.islower():
    lower=True
    break
if not lower:
  print('lower missing ')

upper=False
for i in s:
  if i.isupper():
    upper=True
    break
if not upper:
  print('upper missing ')

digit=False
for i in s:
  if i.isdigit():
    digit=True
    break
if not digit:
  print('digit missing ')

charac=False
sp_char=['_','$','#','@']
for i in s:
  if i in sp_char:
    charac=True
    break
if not charac:
  print('character missing ')

if lower==True & upper==True & digit==True & charac==True:
  print('OK')
```

```python
myLst=[]
flag=True
while flag:
  m=input("Enter total no. in list ")
  if m!='Stop':
    m_Lst=[]
    for i in range(int(m)):
      s=input("enter value in list: ")
      m_Lst.append(int(s))
    myLst.append(m_Lst)
  else:
    flag=False

signal=False
for lst in myLst:
  n=len(lst)
  for j in range(n-1):
    if abs(lst[j+1]-lst[j])<n:
      signal=True
    else:
      signal=False
      break
  if signal:
    print("UB Jumper")
  else:
    print("Not UB Jumper")
```
