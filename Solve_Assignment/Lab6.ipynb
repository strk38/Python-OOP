```python
class Student:
  id=0  # Class variable
  def __init__(self, name, dept, age, cg):
    self.name=name
    self.dept=dept
    self.age=age
    self.cg=cg
    Student.id+=1
  def get_details(self):
    print(self.id)
    print(self.name)
    print(self.dept)
    print(self.age)
    print(self.cg)

  @classmethod
  def from_String(cls, string):
    parts=string.split('-')
    if len(parts)==4:
      name,dept,age,cg=parts
      return cls(name, dept, int(age), float(cg))
    else:
      raise ValueError("Invalid")

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
```

```python
class Assassin:
  total=0
  def __init__(self,name,rate) :
    self.name=name
    self.rate=rate
    Assassin.total+=1

  def printDetails(self):
    print(self.name)
    print(self.rate)
    print(self.total)

  @classmethod
  def failureRate(cls,name,ngRate):
    return cls(name,100-ngRate)

  @classmethod
  def failurePercentage(cls,name,ngRate):
    return cls(name,100-ngRate)


john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
class Passenger:
  count=0
  def __init__(self,name):
    self.name=name
    Passenger.count+=1

  def set_bag_weight(self,wg):
    self.wg=wg
    if wg<=20:
      self.fare=450
    elif 21<=wg<=50:
      self.fare=500
    else:
      self.fare=550

  def printDetail(self):
    print(self.name)
    #print(self.wg)
    print(self.fare)

print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print('Total Passenger:', Passenger.count)
```

```python
class Travel:
  count=0
  def __init__(self,src,to):
    self.src=src
    self.to=to
    self.time=1
    Travel.count+=1

  def display_travel_info(self):
    return '{self.src}\n{self.to}\n{self.time}'

  def set_time(self,time):
    self.time=time

  def set_destination(self,destin):
    self.to=destin

  def set_source(self,source):
    self.src=source

print('No. of Traveller =', Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print('No. of Traveller =', Travel.count)
```

```python
class Employee:

  def __init__(self,nme,wrkP):
    self.name=nme  #instance variable
    self.workingPeriod=wrkP  #instance variable

  @classmethod
  def employeeByJoiningYear(cls, nameN, year):
    return cls(nameN,2021-year)

  @staticmethod
  def experienceCheck(wP, gender):
    if gender=='male':
      if wP<3:
        return 'He is not experienced'
      else:
        return 'He is experienced'
    else:
      if wP<3:
        return 'She is not experienced'
      else:
        return 'She is experienced'

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
class Cat:
  Number_of_cats=0
  def __init__(self,color, verb):
    self.color=color
    self.verb=verb
    Cat.Number_of_cats+=1

  @classmethod
  def no_parameter(cls):
    return cls('White','sitting')
  @classmethod
  def first_parameter(cls,colr):
    return cls(colr,'sitting')
  @classmethod
  def second_parameter(cls,vrb):
    return cls('Grey',vrb)

  def changeColor(self,crl):
    self.color=crl

  def printCat(self):
    print(f'{self.color} cat is {self.verb}')

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
class Cylinder:
  radius=5
  height=18

  def __init__(self,r,h):
    print(f'Deafult radius: {Cylinder.radius} Heighr: {Cylinder.height}')
    Cylinder.radius=r
    Cylinder.height=h
    print(f'Updated radius: {Cylinder.radius} Heighr: {Cylinder.height}')

  @staticmethod
  def area(rad,hei):
    print(f'Area: {2*3.1416*float(rad)*float(rad+hei)}')

  @staticmethod
  def volume(rad,hei):
    print(f'Volume: {3.1416*float(rad)*float(rad)*float(hei)}')

  @classmethod
  def swap(cls,h,r):
    return cls(r,h)

  @classmethod
  def changeFormat(cls,frm):
    f=frm.split('-')
    r1,h1=f
    return cls(r1,h1)


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
