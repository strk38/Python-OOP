```python
class Vehicle:
  def __init__(self):
    self.x = 0
    self.y = 0
  def moveUp(self):
    self.y+=1
  def moveDown(self):
    self.y-=1
  def moveRight(self):
    self.x+=1
  def moveLeft(self):
   self.x-=1
  def __str__(self):
    return '('+str(self.x)+' , '+str(self.y)+')'
#write your code here
class Vehicle2010(Vehicle):
  def __init__(self):
    Vehicle.__init__(self)
  def moveLowerLeft(self):
    self.y-=1
    self.x-=1
  def equals(self,other):
    return self.x==other.x and self.y==other.y

print('Part 1')
print('------')
car = Vehicle()
print(car)
car.moveUp()
print(car)
car.moveLeft()
print(car)
car.moveDown()
print(car)
car.moveRight()
print(car)
print('------')
print('Part 2')
print('------')
car1 = Vehicle2010()
print(car1)
car1.moveLowerLeft()
print(car1)
car2 = Vehicle2010()
car2.moveLeft()
print(car1.equals(car2))
car2.moveDown()
print(car1.equals(car2))
```

```python
class Tournament:
  def __init__(self,name='Default'):
    self.__name = name
  def set_name(self,name):
    self.__name = name
  def get_name(self):
    return self.__name
  def detail(self):
    return f'TournamentName {self.__name}'

class Cricket_Tournament(Tournament):
  def __init__(self,name='Default',team=0,type='No type'):
    Tournament.__init__(self,name)
    self.__team=team
    self.__type=type

  def detail(self):
    return f'{super().detail()}, #of team {self.__team}, type: {self.__type}'

class Tennis_Tournament(Tournament):
  def __init__(self,name='Default',team=0):
    Tournament.__init__(self,name)
    self.__team=team

  def detail(self):
    return f'{super().detail()}, #of team {self.__team}'

#write your code here
ct1 = Cricket_Tournament()
print(ct1.detail())
print("-----------------------")
ct2 = Cricket_Tournament("IPL",10,"t20")
print(ct2.detail())
print("-----------------------")
tt = Tennis_Tournament("Roland Garros",128)
print(tt.detail())
```

```python
class Shape:
  def __init__(self, name='Default', height=0, base=0):
    self.area = 0
    self.name = name
    self.height = height
    self.base = base
  def get_height_base(self):
    return "Height: "+str(self.height)+",Base: "+str(self.base)

#write your code here
class triangle(Shape):
  def __init__(self,name='Default',h=0,b=0):
    super().__init__(name,h,b)
  def calcArea(self):
    self.area=0.5*self.height*self.base
  def printDetail(self):
    return f'{self.name}\n {super().get_height_base()}\n Area: {self.area}'
class trapezoid(Shape):
  def __init__(self,name='Default',h=0,b=0,s=0):
    super().__init__(name,h,b)
    self.__s=s
  def calcArea(self):
    self.area=0.5*self.height*(self.base+self.__s)

  def get_height_base(self):
    return super().get_height_base()+" Side "+str(self.__s)

  def printDetail(self):
    return f'{self.name}\n {self.get_height_base()}\n Area: {self.area}'

tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```

```python
class Football:
  def __init__(self, team_name, name, role):
    self.__team = team_name
    self.__name = name
    self.role = role
    self.earning_per_match = 0
  def get_name_team(self):
    return 'Name: '+self.__name+', Team Name: ' +self.__team
#write your code here
class Player(Football):
  def __init__(self,team_name, name, role,goal,Tot_match):
    super().__init__(team_name,name,role)
    self.goal=goal
    self.tot_match=Tot_match

  def calculate_ratio(self):
    self.ratio= self.goal/self.tot_match

  def calculate_earning(self):
    return self.goal*1000+10*self.tot_match

  def print_details(self):
    print(f'{self.get_name_team()}\n {self.role}\n {self.goal} \n {self.ratio} \n {self.calculate_earning()}k')

class Manager(Football):
  def __init__(self,team,name,role,mW):
    super().__init__(team,name,role)
    self.mW=mW

  def calculate_earning(self):
    return self.mW*1000

  def print_details(self):
    print(f'{self.get_name_team()}\n {self.role}\n {self.mW} \n {self.calculate_earning()}k')

player_one = Player('Juventus', 'Ronaldo', 'Striker', 25, 32)
player_one.calculate_ratio()
player_one.print_details()
print('------------------------------------------')
manager_one = Manager('Real Madrid', 'Zidane', 'Manager', 25)
manager_one.print_details()
```
